```python
# Cell 1: Install required packages
!pip install segmentation-models-pytorch 
!pip install albumentations
!pip install torchvision
!pip install opencv-python
!pip install -U git+https://github.com/rwightman/pytorch-image-models.git

print("All packages installed successfully!")

# Cell 2: Performance Optimizations
import torch
torch.backends.cudnn.benchmark = True  # Enable cuDNN auto-tuner
torch.backends.cuda.matmul.allow_tf32 = True  # Enable TF32 for faster matrix math
torch.backends.cudnn.allow_tf32 = True

import os
os.environ["CUDA_LAUNCH_BLOCKING"] = "0"  # Disable for faster async execution

print("⚡ Performance optimizations enabled!")

# Cell 3: Imports & Optimized Configuration
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import cv2
import albumentations as A
from albumentations.pytorch import ToTensorV2
import segmentation_models_pytorch as smp
import matplotlib.pyplot as plt
from tqdm import tqdm
import torch.cuda.amp as amp

# OPTIMIZED Configuration for Speed
class FastConfig:
    # Model - Using lighter backbone
    MODEL_NAME = "DeepLabV3Plus"
    ENCODER = "efficientnet-b3"  # Much faster than B5
    ENCODER_WEIGHTS = "imagenet"
    CLASSES = 7
    IN_CHANNELS = 3
    
    # Training - OPTIMIZED for speed
    DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
    BATCH_SIZE = 16  # Increased batch size
    LEARNING_RATE = 2e-4  # Slightly higher
    WEIGHT_DECAY = 1e-4
    EPOCHS = 10
    IMAGE_SIZE = 256  # Reduced from 512 (faster processing)
    
    # Early stopping - ONLY based on patience, no accuracy cap
    PATIENCE = 15  # Number of epochs to wait before stopping
    
    # Data Paths
    DATA_PATH = "/kaggle/input/sen-lulc/SEN-2 LULC"
    TRAIN_IMAGES = os.path.join(DATA_PATH, "train_images/train")
    TRAIN_MASKS = os.path.join(DATA_PATH, "train_masks/train")
    VAL_IMAGES = os.path.join(DATA_PATH, "val_images/val")
    VAL_MASKS = os.path.join(DATA_PATH, "val_masks/val")
    TEST_IMAGES = os.path.join(DATA_PATH, "test_images/test")
    TEST_MASKS = os.path.join(DATA_PATH, "test_masks/test")
    
config = FastConfig()
print(f"⚡ Using device: {config.DEVICE}")
print(f"📊 Optimized config: Batch Size={config.BATCH_SIZE}, Image Size={config.IMAGE_SIZE}")
print(f"🛑 Early stopping: Patience={config.PATIENCE} epochs without improvement")
print(f"🎯 Training will continue to {config.EPOCHS} epochs unless no improvement for {config.PATIENCE} epochs")

# Verify paths
print("Checking paths...")
print(f"Train images: {os.path.exists(config.TRAIN_IMAGES)}")
print(f"Train masks: {os.path.exists(config.TRAIN_MASKS)}")

# Cell 4: Optimized Dataset Class
class FastSatelliteDataset(Dataset):
    def __init__(self, images_path, masks_path, transform=None, num_classes=7):
        self.images_path = images_path
        self.masks_path = masks_path
        self.transform = transform
        self.num_classes = num_classes
        
        # Get files - assuming same names for images and masks
        self.images = sorted([f for f in os.listdir(images_path) if f.endswith(('.jpg', '.png', '.jpeg', '.tif'))])
        self.masks = sorted([f for f in os.listdir(masks_path) if f.endswith(('.jpg', '.png', '.jpeg', '.tif'))])
        
        assert len(self.images) == len(self.masks), f"Mismatch: {len(self.images)} images vs {len(self.masks)} masks"
        print(f"📁 Found {len(self.images)} samples in {os.path.basename(images_path)}")
        
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        # Read image and mask
        image_path = os.path.join(self.images_path, self.images[idx])
        mask_path = os.path.join(self.masks_path, self.masks[idx])
        
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        
        # Quick mask value adjustment (assuming 0-6 range to 0-4)
        if mask is not None:
            mask = np.clip(mask, 0, self.num_classes - 1)
        
        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']
            mask = augmented['mask']
        
        return image, mask.long()

# SIMPLIFIED transforms for speed
def get_fast_train_transform():
    return A.Compose([
        A.Resize(config.IMAGE_SIZE, config.IMAGE_SIZE),
        A.HorizontalFlip(p=0.3),  # Reduced probability for speed
        A.VerticalFlip(p=0.3),
        A.RandomBrightnessContrast(p=0.2),  # Fewer augmentations
        A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ToTensorV2(),
    ])

def get_fast_val_transform():
    return A.Compose([
        A.Resize(config.IMAGE_SIZE, config.IMAGE_SIZE),
        A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
        ToTensorV2(),
    ])

# Cell 5: Model & Simple Loss
def create_fast_model():
    model = smp.DeepLabV3Plus(
        encoder_name=config.ENCODER,
        encoder_weights=config.ENCODER_WEIGHTS,
        in_channels=config.IN_CHANNELS,
        classes=config.CLASSES,
    )
    return model

# SIMPLE Loss for speed
class FastLoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.ce_loss = nn.CrossEntropyLoss()
        
    def forward(self, pred, target):
        return self.ce_loss(pred, target)

# Enhanced Metrics with IoU and Dice per class
def calculate_iou_per_class(preds, targets, num_classes, epsilon=1e-6):
    """
    Calculate IoU for each class
    """
    preds = torch.argmax(preds, dim=1)
    ious = []
    
    for cls in range(num_classes):
        pred_cls = (preds == cls)
        target_cls = (targets == cls)
        
        intersection = (pred_cls & target_cls).sum().float()
        union = (pred_cls | target_cls).sum().float()
        
        iou = (intersection + epsilon) / (union + epsilon)
        ious.append(iou.item())
    
    return ious

def calculate_dice_per_class(preds, targets, num_classes, epsilon=1e-6):
    """
    Calculate Dice coefficient for each class
    """
    preds = torch.argmax(preds, dim=1)
    dice_scores = []
    
    for cls in range(num_classes):
        pred_cls = (preds == cls)
        target_cls = (targets == cls)
        
        intersection = (pred_cls & target_cls).sum().float()
        dice = (2. * intersection + epsilon) / (pred_cls.sum() + target_cls.sum() + epsilon)
        dice_scores.append(dice.item())
    
    return dice_scores

def calculate_mean_iou(preds, targets, num_classes, epsilon=1e-6):
    """
    Calculate mean IoU across all classes
    """
    ious = calculate_iou_per_class(preds, targets, num_classes, epsilon)
    return np.mean(ious)

def calculate_mean_dice(preds, targets, num_classes, epsilon=1e-6):
    """
    Calculate mean Dice coefficient across all classes
    """
    dice_scores = calculate_dice_per_class(preds, targets, num_classes, epsilon)
    return np.mean(dice_scores)

def calculate_accuracy(preds, targets):
    preds = torch.argmax(preds, dim=1)
    correct = (preds == targets).sum().float()
    total = targets.numel()
    return (correct / total).item()

# Cell 6: Optimized DataLoader Setup
print("🚀 Creating optimized datasets...")

train_dataset = FastSatelliteDataset(
    images_path=config.TRAIN_IMAGES,
    masks_path=config.TRAIN_MASKS,
    transform=get_fast_train_transform(),
    num_classes=config.CLASSES
)

val_dataset = FastSatelliteDataset(
    images_path=config.VAL_IMAGES,
    masks_path=config.VAL_MASKS,
    transform=get_fast_val_transform(),
    num_classes=config.CLASSES
)

# OPTIMIZED DataLoaders
train_loader = DataLoader(
    train_dataset,
    batch_size=config.BATCH_SIZE,
    shuffle=True,
    num_workers=2,  # Reduced workers
    pin_memory=True,
    persistent_workers=False,  # Disable for speed
    drop_last=True  # Drop incomplete batches
)

val_loader = DataLoader(
    val_dataset,
    batch_size=config.BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    pin_memory=True,
    persistent_workers=False
)

print(f"✅ Train samples: {len(train_dataset)}")
print(f"✅ Val samples: {len(val_dataset)}")
print(f"✅ Train batches: {len(train_loader)}")
print(f"✅ Val batches: {len(val_loader)}")

# Quick test
for images, masks in train_loader:
    print(f"🎯 Test batch - Images: {images.shape}, Masks: {masks.shape}")
    break

# Cell 7: Fast Training Loop with IoU and Dice Metrics
class FastTrainer:
    def __init__(self, model, train_loader, val_loader, criterion, optimizer, scheduler, device):
        self.model = model
        self.train_loader = train_loader
        self.val_loader = val_loader
        self.criterion = criterion
        self.optimizer = optimizer
        self.scheduler = scheduler
        self.device = device
        self.scaler = amp.GradScaler()
        
        self.train_losses = []
        self.val_losses = []
        self.train_accuracies = []
        self.val_accuracies = []
        self.train_ious = []
        self.val_ious = []
        self.train_dices = []
        self.val_dices = []
        self.train_ious_per_class = []
        self.val_ious_per_class = []
        self.train_dices_per_class = []
        self.val_dices_per_class = []
        
    def train_epoch(self):
        self.model.train()
        running_loss = 0.0
        running_accuracy = 0.0
        running_iou = 0.0
        running_dice = 0.0
        
        # For per-class metrics
        running_ious_per_class = np.zeros(config.CLASSES)
        running_dices_per_class = np.zeros(config.CLASSES)
        
        # FAST progress bar - no per-batch metrics
        pbar = tqdm(self.train_loader, desc="Training")
        for images, masks in pbar:
            images = images.to(self.device, non_blocking=True)
            masks = masks.to(self.device, non_blocking=True)
            
            self.optimizer.zero_grad()
            
            # Mixed precision training
            with amp.autocast():
                outputs = self.model(images)
                loss = self.criterion(outputs, masks)
            
            self.scaler.scale(loss).backward()
            self.scaler.step(self.optimizer)
            self.scaler.update()
            
            # Calculate metrics
            accuracy = calculate_accuracy(outputs, masks)
            batch_iou = calculate_mean_iou(outputs, masks, config.CLASSES)
            batch_dice = calculate_mean_dice(outputs, masks, config.CLASSES)
            batch_ious_per_class = calculate_iou_per_class(outputs, masks, config.CLASSES)
            batch_dices_per_class = calculate_dice_per_class(outputs, masks, config.CLASSES)
            
            running_loss += loss.item()
            running_accuracy += accuracy
            running_iou += batch_iou
            running_dice += batch_dice
            running_ious_per_class += np.array(batch_ious_per_class)
            running_dices_per_class += np.array(batch_dices_per_class)
            
            # Update progress bar less frequently
            if len(pbar) % 10 == 0:
                pbar.set_postfix({'Loss': f'{loss.item():.4f}', 'Acc': f'{accuracy:.4f}'})
        
        epoch_loss = running_loss / len(self.train_loader)
        epoch_accuracy = running_accuracy / len(self.train_loader)
        epoch_iou = running_iou / len(self.train_loader)
        epoch_dice = running_dice / len(self.train_loader)
        epoch_ious_per_class = running_ious_per_class / len(self.train_loader)
        epoch_dices_per_class = running_dices_per_class / len(self.train_loader)
        
        return (epoch_loss, epoch_accuracy, epoch_iou, epoch_dice, 
                epoch_ious_per_class, epoch_dices_per_class)
    
    def validate_epoch(self):
        self.model.eval()
        running_loss = 0.0
        running_accuracy = 0.0
        running_iou = 0.0
        running_dice = 0.0
        
        # For per-class metrics
        running_ious_per_class = np.zeros(config.CLASSES)
        running_dices_per_class = np.zeros(config.CLASSES)
        
        with torch.no_grad():
            for images, masks in self.val_loader:
                images = images.to(self.device, non_blocking=True)
                masks = masks.to(self.device, non_blocking=True)
                
                with amp.autocast():
                    outputs = self.model(images)
                    loss = self.criterion(outputs, masks)
                
                # Calculate metrics
                accuracy = calculate_accuracy(outputs, masks)
                batch_iou = calculate_mean_iou(outputs, masks, config.CLASSES)
                batch_dice = calculate_mean_dice(outputs, masks, config.CLASSES)
                batch_ious_per_class = calculate_iou_per_class(outputs, masks, config.CLASSES)
                batch_dices_per_class = calculate_dice_per_class(outputs, masks, config.CLASSES)
                
                running_loss += loss.item()
                running_accuracy += accuracy
                running_iou += batch_iou
                running_dice += batch_dice
                running_ious_per_class += np.array(batch_ious_per_class)
                running_dices_per_class += np.array(batch_dices_per_class)
        
        epoch_loss = running_loss / len(self.val_loader)
        epoch_accuracy = running_accuracy / len(self.val_loader)
        epoch_iou = running_iou / len(self.val_loader)
        epoch_dice = running_dice / len(self.val_loader)
        epoch_ious_per_class = running_ious_per_class / len(self.val_loader)
        epoch_dices_per_class = running_dices_per_class / len(self.val_loader)
        
        return (epoch_loss, epoch_accuracy, epoch_iou, epoch_dice, 
                epoch_ious_per_class, epoch_dices_per_class)
    
    def print_class_metrics(self, ious_per_class, dices_per_class, prefix=""):
        """Print IoU and Dice metrics for each class"""
        print(f"{prefix}Per-Class Metrics:")
        for cls in range(config.CLASSES):
            print(f"  Class {cls}: IoU={ious_per_class[cls]:.4f}, Dice={dices_per_class[cls]:.4f}")
    
    def train(self, epochs):
        print("🚀 Starting FAST training...")
        best_accuracy = 0.0
        patience_counter = 0
        best_epoch = 0
        
        for epoch in range(epochs):
            print(f"\n📈 Epoch {epoch+1}/{epochs}")
            print("=" * 60)
            
            # Train
            (train_loss, train_acc, train_iou, train_dice, 
             train_ious_per_class, train_dices_per_class) = self.train_epoch()
            
            # Validate
            (val_loss, val_acc, val_iou, val_dice, 
             val_ious_per_class, val_dices_per_class) = self.validate_epoch()
            
            # Update scheduler
            self.scheduler.step()
            
            # Store metrics
            self.train_losses.append(train_loss)
            self.val_losses.append(val_loss)
            self.train_accuracies.append(train_acc)
            self.val_accuracies.append(val_acc)
            self.train_ious.append(train_iou)
            self.val_ious.append(val_iou)
            self.train_dices.append(train_dice)
            self.val_dices.append(val_dice)
            self.train_ious_per_class.append(train_ious_per_class)
            self.val_ious_per_class.append(val_ious_per_class)
            self.train_dices_per_class.append(train_dices_per_class)
            self.val_dices_per_class.append(val_dices_per_class)
            
            # Print epoch results
            print(f"\n📊 Epoch {epoch+1} Results:")
            print(f"Train - Loss: {train_loss:.4f}, Acc: {train_acc:.4f}, mIoU: {train_iou:.4f}, mDice: {train_dice:.4f}")
            print(f"Val   - Loss: {val_loss:.4f}, Acc: {val_acc:.4f}, mIoU: {val_iou:.4f}, mDice: {val_dice:.4f}")
            
            # Print per-class metrics
            self.print_class_metrics(train_ious_per_class, train_dices_per_class, "Train")
            self.print_class_metrics(val_ious_per_class, val_dices_per_class, "Val  ")
            
            # Save best model and check for improvement
            if val_acc > best_accuracy:
                best_accuracy = val_acc
                best_epoch = epoch
                torch.save({
                    'epoch': epoch,
                    'model_state_dict': self.model.state_dict(),
                    'optimizer_state_dict': self.optimizer.state_dict(),
                    'best_accuracy': best_accuracy,
                    'train_loss': train_loss,
                    'val_loss': val_loss,
                    'val_iou': val_iou,
                    'val_dice': val_dice,
                    'val_ious_per_class': val_ious_per_class,
                    'val_dices_per_class': val_dices_per_class,
                }, 'best_model.pth')
                print(f"\n💾 New best model saved! Accuracy: {best_accuracy:.4f}")
                patience_counter = 0  # Reset patience counter
            else:
                patience_counter += 1
                print(f"\n🔄 No improvement for {patience_counter} epoch(s)")
            
            # Early stopping if no improvement for 'patience' epochs
            if patience_counter >= config.PATIENCE:
                print(f"🛑 Early stopping after {config.PATIENCE} epochs without improvement")
                print(f"🏆 Best accuracy: {best_accuracy:.4f} at epoch {best_epoch + 1}")
                break
            
            # Learning rate info
            current_lr = self.scheduler.get_last_lr()[0]
            print(f"📚 Learning Rate: {current_lr:.2e}")
            
            # Check if this is the last epoch
            if epoch == epochs - 1:
                print(f"\n🏁 Completed all {epochs} epochs!")
        
        # Final summary
        print("\n" + "="*80)
        print("🏁 TRAINING COMPLETED!")
        print(f"🏆 Best Validation Accuracy: {best_accuracy:.4f}")
        print(f"📈 Achieved at epoch: {best_epoch + 1}")
        print(f"📊 Final Training Accuracy: {self.train_accuracies[-1]:.4f}")
        print(f"📊 Final Validation Accuracy: {self.val_accuracies[-1]:.4f}")
        print(f"🎯 Final Training mIoU: {self.train_ious[-1]:.4f}")
        print(f"🎯 Final Validation mIoU: {self.val_ious[-1]:.4f}")
        print(f"🎯 Final Training mDice: {self.train_dices[-1]:.4f}")
        print(f"🎯 Final Validation mDice: {self.val_dices[-1]:.4f}")
        print(f"📉 Total Epochs Trained: {len(self.train_accuracies)}")
        print("="*80)
        
        return best_accuracy

# Cell 8: Initialize and Start Fast Training
# Create model
model = create_fast_model()
model = model.to(config.DEVICE)

# Simple loss and optimizer
criterion = FastLoss()
optimizer = torch.optim.AdamW(
    model.parameters(), 
    lr=config.LEARNING_RATE, 
    weight_decay=config.WEIGHT_DECAY
)

# Simple scheduler
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=20, gamma=0.5)

# Create trainer
trainer = FastTrainer(
    model=model,
    train_loader=train_loader,
    val_loader=val_loader,
    criterion=criterion,
    optimizer=optimizer,
    scheduler=scheduler,
    device=config.DEVICE
)

print("🔥 Starting training with optimizations...")
print(f"📋 Training for {config.EPOCHS} epochs maximum")
print(f"🛑 Early stopping only if no improvement for {config.PATIENCE} consecutive epochs")
print("🎯 No accuracy cap - model can reach 100% if possible!")
print(f"📊 Will display IoU and Dice coefficients per class every epoch")

best_accuracy = trainer.train(config.EPOCHS)
print(f"🎉 Final Best Accuracy: {best_accuracy:.4f}")

# Cell 9: Enhanced Results Visualization
plt.figure(figsize=(20, 12))

# Loss
plt.subplot(2, 3, 1)
plt.plot(trainer.train_losses, label='Train Loss', linewidth=2)
plt.plot(trainer.val_losses, label='Val Loss', linewidth=2)
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, alpha=0.3)

# Accuracy
plt.subplot(2, 3, 2)
plt.plot(trainer.train_accuracies, label='Train Accuracy', linewidth=2)
plt.plot(trainer.val_accuracies, label='Val Accuracy', linewidth=2)
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True, alpha=0.3)

# IoU
plt.subplot(2, 3, 3)
plt.plot(trainer.train_ious, label='Train mIoU', linewidth=2)
plt.plot(trainer.val_ious, label='Val mIoU', linewidth=2)
plt.title('Training and Validation mIoU')
plt.xlabel('Epoch')
plt.ylabel('mIoU')
plt.legend()
plt.grid(True, alpha=0.3)

# Dice
plt.subplot(2, 3, 4)
plt.plot(trainer.train_dices, label='Train mDice', linewidth=2)
plt.plot(trainer.val_dices, label='Val mDice', linewidth=2)
plt.title('Training and Validation mDice')
plt.xlabel('Epoch')
plt.ylabel('mDice')
plt.legend()
plt.grid(True, alpha=0.3)

# Per-class IoU (last epoch)
plt.subplot(2, 3, 5)
classes = range(config.CLASSES)
train_ious_last = trainer.train_ious_per_class[-1]
val_ious_last = trainer.val_ious_per_class[-1]
x = np.arange(len(classes))
width = 0.35
plt.bar(x - width/2, train_ious_last, width, label='Train IoU', alpha=0.7)
plt.bar(x + width/2, val_ious_last, width, label='Val IoU', alpha=0.7)
plt.title('Per-Class IoU (Last Epoch)')
plt.xlabel('Class')
plt.ylabel('IoU')
plt.xticks(x, classes)
plt.legend()
plt.grid(True, alpha=0.3)

# Per-class Dice (last epoch)
plt.subplot(2, 3, 6)
train_dices_last = trainer.train_dices_per_class[-1]
val_dices_last = trainer.val_dices_per_class[-1]
plt.bar(x - width/2, train_dices_last, width, label='Train Dice', alpha=0.7)
plt.bar(x + width/2, val_dices_last, width, label='Val Dice', alpha=0.7)
plt.title('Per-Class Dice (Last Epoch)')
plt.xlabel('Class')
plt.ylabel('Dice')
plt.xticks(x, classes)
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print final metrics
print("\n📊 FINAL METRICS SUMMARY:")
print(f"🏆 Best Validation Accuracy: {best_accuracy:.4f}")
print(f"📊 Final Training Accuracy: {trainer.train_accuracies[-1]:.4f}")
print(f"📊 Final Validation Accuracy: {trainer.val_accuracies[-1]:.4f}")
print(f"🎯 Final Training mIoU: {trainer.train_ious[-1]:.4f}")
print(f"🎯 Final Validation mIoU: {trainer.val_ious[-1]:.4f}")
print(f"🎯 Final Training mDice: {trainer.train_dices[-1]:.4f}")
print(f"🎯 Final Validation mDice: {trainer.val_dices[-1]:.4f}")
print(f"📈 Total Epochs Completed: {len(trainer.train_accuracies)}")

# Print final per-class metrics
print("\n🎯 FINAL PER-CLASS METRICS (Validation):")
for cls in range(config.CLASSES):
    print(f"  Class {cls}: IoU={trainer.val_ious_per_class[-1][cls]:.4f}, Dice={trainer.val_dices_per_class[-1][cls]:.4f}")

# Cell 10: Enhanced Test Function with IoU and Dice
def quick_test():
    print("\n🧪 Running Enhanced Test with IoU and Dice...")
    
    # Load best model
    try:
        checkpoint = torch.load('best_model.pth')
        model.load_state_dict(checkpoint['model_state_dict'])
        model.eval()
        print("✅ Best model loaded successfully!")
    except Exception as e:
        print(f"❌ Error loading model: {e}")
        return
    
    if os.path.exists(config.TEST_IMAGES) and os.path.exists(config.TEST_MASKS):
        test_dataset = FastSatelliteDataset(
            images_path=config.TEST_IMAGES,
            masks_path=config.TEST_MASKS,
            transform=get_fast_val_transform(),
            num_classes=config.CLASSES
        )
        
        test_loader = DataLoader(test_dataset, batch_size=config.BATCH_SIZE, shuffle=False)
        
        test_accuracy = 0.0
        test_loss = 0.0
        test_iou = 0.0
        test_dice = 0.0
        test_ious_per_class = np.zeros(config.CLASSES)
        test_dices_per_class = np.zeros(config.CLASSES)
        
        print(f"🧪 Testing on {len(test_dataset)} samples...")
        
        with torch.no_grad():
            for images, masks in tqdm(test_loader, desc="Testing"):
                images = images.to(config.DEVICE)
                masks = masks.to(config.DEVICE)
                outputs = model(images)
                
                # Calculate loss
                loss = criterion(outputs, masks)
                test_loss += loss.item()
                
                # Calculate metrics
                test_accuracy += calculate_accuracy(outputs, masks)
                test_iou += calculate_mean_iou(outputs, masks, config.CLASSES)
                test_dice += calculate_mean_dice(outputs, masks, config.CLASSES)
                
                batch_ious_per_class = calculate_iou_per_class(outputs, masks, config.CLASSES)
                batch_dices_per_class = calculate_dice_per_class(outputs, masks, config.CLASSES)
                test_ious_per_class += np.array(batch_ious_per_class)
                test_dices_per_class += np.array(batch_dices_per_class)
        
        test_accuracy /= len(test_loader)
        test_loss /= len(test_loader)
        test_iou /= len(test_loader)
        test_dice /= len(test_loader)
        test_ious_per_class /= len(test_loader)
        test_dices_per_class /= len(test_loader)
        
        print(f"\n🧪 TEST RESULTS:")
        print(f"   Loss: {test_loss:.4f}")
        print(f"   Accuracy: {test_accuracy:.4f}")
        print(f"   mIoU: {test_iou:.4f}")
        print(f"   mDice: {test_dice:.4f}")
        print(f"   Samples: {len(test_dataset)}")
        
        print(f"\n🧪 TEST PER-CLASS METRICS:")
        for cls in range(config.CLASSES):
            print(f"   Class {cls}: IoU={test_ious_per_class[cls]:.4f}, Dice={test_dices_per_class[cls]:.4f}")
    else:
        print("❌ Test dataset not found")

# Run enhanced test
quick_test()

# Cell 11: Save Enhanced Training History
def save_training_history(trainer, filename='training_history.pth'):
    history = {
        'train_losses': trainer.train_losses,
        'val_losses': trainer.val_losses,
        'train_accuracies': trainer.train_accuracies,
        'val_accuracies': trainer.val_accuracies,
        'train_ious': trainer.train_ious,
        'val_ious': trainer.val_ious,
        'train_dices': trainer.train_dices,
        'val_dices': trainer.val_dices,
        'train_ious_per_class': trainer.train_ious_per_class,
        'val_ious_per_class': trainer.val_ious_per_class,
        'train_dices_per_class': trainer.train_dices_per_class,
        'val_dices_per_class': trainer.val_dices_per_class,
        'best_accuracy': best_accuracy,
        'config': {
            'EPOCHS': config.EPOCHS,
            'BATCH_SIZE': config.BATCH_SIZE,
            'LEARNING_RATE': config.LEARNING_RATE,
            'IMAGE_SIZE': config.IMAGE_SIZE,
            'PATIENCE': config.PATIENCE,
            'CLASSES': config.CLASSES
        }
    }
    
    torch.save(history, filename)
    print(f"📁 Enhanced training history saved to {filename}")

save_training_history(trainer)

print("\n" + "="*80)
print("🎉 ALL COMPLETED SUCCESSFULLY!")
print("="*80)
```

```python
# Cell 12: Full Visualization with Legends Inside Plot
import random
from matplotlib.patches import Patch

# Define class labels and colors
CLASS_NAMES = {
    0: "Barren",
    1: "Urban",
    2: "Water",
    3: "Forest",
    4: "Agricultural",
    5: "Green Land",
    6: "Other"
}

CLASS_COLORS = {
    0: ("lightgray", [200, 200, 200]),   # Barren
    1: ("red",       [220, 20, 60]),     # Urban
    2: ("blue",      [30, 144, 255]),    # Water
    3: ("green",     [34, 139, 34]),     # Forest
    4: ("gold",      [255, 215, 0]),     # Agricultural
    5: ("lime",      [0, 255, 127]),     # Green Land
    6: ("violet",    [148, 0, 211])      # Other
}

def colorize_mask(mask):
    """Convert mask (H, W) -> color image (H, W, 3)."""
    color_mask = np.zeros((*mask.shape, 3))
    for cls_id, (_, rgb) in CLASS_COLORS.items():
        color_mask[mask == cls_id] = np.array(rgb) / 255.0
    return color_mask

def visualize_predictions(dataset, model, device, num_samples=3):
    model.eval()
    indices = random.sample(range(len(dataset)), num_samples)
    
    for i, idx in enumerate(indices):
        image, mask = dataset[idx]
        image_tensor = image.unsqueeze(0).to(device)
        
        with torch.no_grad():
            pred = model(image_tensor)
            pred = torch.argmax(pred, dim=1).squeeze().cpu().numpy()
        
        image_np = image.permute(1, 2, 0).cpu().numpy()
        mask_np = mask.cpu().numpy()
        
        # Classes present
        gt_classes = np.unique(mask_np)
        pred_classes = np.unique(pred)
        
        # -------------------------
        # Plot Images & Masks
        # -------------------------
        fig, axs = plt.subplots(1, 3, figsize=(15, 5))
        
        axs[0].imshow(image_np)
        axs[0].set_title("Original Image")
        axs[0].axis("off")
        
        axs[1].imshow(colorize_mask(mask_np))
        axs[1].set_title("Ground Truth Mask")
        axs[1].axis("off")
        
        axs[2].imshow(colorize_mask(pred))
        axs[2].set_title("Predicted Mask")
        axs[2].axis("off")
        
        fig.suptitle(f"🖼️ Image {i+1} (Index {idx})", fontsize=14, fontweight="bold")
        
        # -------------------------
        # Add Class Legend
        # -------------------------
        legend_elements = []
        for cls_id in np.unique(np.concatenate([gt_classes, pred_classes])):
            cname = CLASS_NAMES[cls_id]
            color_name, _ = CLASS_COLORS[cls_id]
            legend_elements.append(Patch(facecolor=color_name, edgecolor='black', label=f"{cname} ({color_name})"))
        
        fig.legend(handles=legend_elements, loc="lower center", ncol=3, fontsize=10, frameon=False)
        
        plt.tight_layout(rect=[0, 0.05, 1, 1])  # leave space for legend
        plt.show()

# Run visualization
visualize_predictions(val_dataset, model, config.DEVICE, num_samples=3)
```
